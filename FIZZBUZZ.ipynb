```
#normal way to do fizz buzz problem for 1 to 100 simply------------
i=1
while i <= 100:
    if i%3==0:
        print("Fizz", end="")
        if i%5==0:
            print("Buzz", end="")
    elif i%5==0:
        print("Buzz", end="")
    else:
        print(i, end="")
    print()
    i+=1
```

```
#fizz buzz problem where 1st line of input is reffered to the number of testcases(T) & in another line of input it takes T ranges:N (starting from 1 to N):---
"""
sample input:------
2
3 15

Sample output:-----
1
2
Fizz
1
2
Fizz
4
Buzz
Fizz
7
8
Fizz
Buzz
11
Fizz
13
14
FizzBuzz
""" 
T=int(input())
N =list(map(int,input().split()))[:T]
  
# printing the list using loop 
for x in range(len(N)): 
    i=1
    while i <= N[x]:
      if i%3==0:
         print("Fizz", end="")
         if i%5==0:
            print("Buzz", end="")
      elif i%5==0:
         print("Buzz", end="")
      else:
         print(i, end="")
      print()
      i+=1
    
```
